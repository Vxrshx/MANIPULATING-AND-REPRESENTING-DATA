```python
import json  # Importing the JSON library for the parsing JSON data
from tqdm import tqdm  # Importing tqdm for adding progress bars to loops

def load_json_lines(file_path):
    """Read JSON data from a file line by line, skipping any malformed entries."""
    data = []  # Initialize an empty list to store valid JSON objects

    with open(file_path, "r") as file:  # Open the specified file in read mode
        # Iterate over each line in the file with a progress bar
        for line in tqdm(file, desc=f"Loading {file_path.split('/')[-1]}"):
            try:
                # Attempt to parse the JSON line and append to the data list
                data.append(json.loads(line))
            except json.JSONDecodeError:
                # On a JSON parsing error, skip this line and continue
                continue

    return data  # Return the list of successfully parsed JSON objects

# Load product metadata from the specified JSON file
product_metadata = load_json_lines("/jcpenney_products.json")

# Load reviewer metadata using the same function for consistency
reviewer_metadata = load_json_lines("/jcpenney_reviewers.json")
```

The code imports the `json` and `tqdm` libraries, where `json` is for managing the JSON data  and `tqdm`is used for enhancing loop functionality.It includes a function to load JSON data line by line initializing an empty list called `data` to hold the analyze the data. A context manager is used to open a file in readable  mode allowing individual lines to be iterated over with a progress bar indicating the status. A try except block manages potential "JSON Decode Error" exceptions,

Ensuring the process continues even if a line is not properly formatted in this code .After processing a list of containing successfully analyzed JSON data is produced.The function is used first to extract product details from "jcpenney_products.json," followed by reviewer details from "jcpenney_reviewers.json."

```python
def enhanced_missing_analysis(df, title):
    """Perform comprehensive missing value analysis with visualization.
    Provides both visual heatmap and statistical summary of missing values
    in the input DataFrame.
    Args:
        df (pd.DataFrame): Input DataFrame to analyze
        title (str): Descriptive name for the dataset (used in plots/titles)
    """
    # Calculate missing value statistics
    missing = df.isnull().sum()  # Raw count of nulls per column
    missing_pct = missing / len(df) * 100

    # Create visualization of missing data pattern
    plt.figure(figsize=(12,6))
    # Generate heatmap where yellow indicates missing values
    sns.heatmap(df.isnull(), cbar=False, cmap='viridis')  # Yellow = missing
    plt.title(f'Missing Values Heatmap: {title}\nTotal Rows: {len(df)}', pad=20)
    plt.show()  # Display the heatmap

    # Create detailed summary table of missing values
    missing_stats = pd.DataFrame({
        'Missing Count': missing,
        'Missing %': missing_pct.round(2),  # Percentage rounded 2 decimal places
        'Data Type': df.dtypes  # Original column data types
    }).sort_values('Missing Count', ascending=False)  # Sort by most missing

    # Display only columns that have missing values
    print(f"\nMissing Value Statistics for {title}:")
    display(missing_stats[missing_stats['Missing Count'] > 0])  # Only show columns with missing values

# Dictionary of datasets to analyse
datasets = {
    "Products CSV": products,  # Main product catalog
    "Reviews CSV": reviews,  # Customer reviews
    "Users CSV": users,  # User demographics
    "Products JSON": product_metadata_df,  # Extended product details
    "Reviewers JSON": reviewer_metadata_df  # Reviewer profiles
}

# Execute missing value analysis for each dataset
for name, df in datasets.items():
    enhanced_missing_analysis(df, name)
```

This code conducts a comprehensive missing values analysis across the datasets using a function called `enhanced_missing_analysis`. This function takes a DataFrame and a descriptive title as parameters, calculates the number and percentage of missing values from each column and visualizes the missing data in the form of a yellow highlights heatmap where one can quickly identify problem areas within the dataset. Finally when it creates a summary table with count, percentage and types of columns with missing values and sorted by the number of missing entries.

It handles multiple datasets such as products, reviews and user demographics, stored in a dictionary and iterates over each entry to apply an analysis function.This will provide a quick way to get visual and statistical insights with respect to data quality and might suggest subsequent steps in terms of cleaning or imputing data. In general, this will enhance the understanding of missing data for effective data management and preparation.

```python
import pandas as pd
import numpy as np

#Safe JSON loader (handles line-delimited JSON too)

def safe_read_json(path):
    """Try loading JSON file normally, then as line-delimited if needed."""
    try:
        return pd.read_json(path)
    except ValueError:
        return pd.read_json(path, lines=True)

#Load your datasets

csv_path = "/products.csv"
json_path = "/jcpenney_products.json"

products_csv = pd.read_csv(csv_path)
products_json = safe_read_json(json_path)

print(" Datasets loaded successfully!")
print("CSV Columns:", list(products_csv.columns))
print("JSON Columns:", list(products_json.columns))

#Ensure 'clean_price' exists in JSON data

if 'clean_price' not in products_json.columns:
    # Try to find a column that likely contains price info
    possible_price_col = next((col for col in products_json.columns if 'price' in col.lower()), None)
    if possible_price_col:
        products_json['clean_price'] = (
            products_json[possible_price_col]
            .astype(str)
            .str.replace(r'[^0-9\.]', '', regex=True)
            .replace('', np.nan)
            .astype(float)
        )
        print(f" Created 'clean_price' from '{possible_price_col}' column.")
    else:
        raise ValueError("JSON file does not contain a recognizable price column!")

#Handle missing JSON prices (imputation)

json_median = products_json['clean_price'].median(skipna=True)
products_json['clean_price'] = products_json['clean_price'].fillna(json_median)

print("\n Missing Values Check After Imputation:")
print("JSON - Missing Prices:", products_json['clean_price'].isna().sum())
print(f"Median used for imputation: {json_median:.2f}")

# 5. Normalize price scale between CSV & JSON
# Detect CSV price column safely
csv_price_col = next((col for col in products_csv.columns if 'price' in col.lower()), None)
if not csv_price_col:
    raise ValueError(" Could not find a price column in CSV!")

csv_mean = products_csv[csv_price_col].mean(skipna=True)
json_mean = products_json['clean_price'].mean(skipna=True)

if json_mean == 0 or np.isnan(json_mean):
    raise ValueError(" JSON mean price is invalid — cannot normalize.")

price_adjustment_factor = csv_mean / json_mean
products_json['adjusted_price'] = products_json['clean_price'] * price_adjustment_factor

print(f"\n Price normalization factor applied: {price_adjustment_factor:.4f}")

# 6. Save cleaned datasets

csv_output = "/content/final_products_csv.csv"
json_output = "/content/final_products_json.csv"

products_csv.to_csv(csv_output, index=False)
products_json.to_csv(json_output, index=False)

print(f"\n Cleaned CSV saved to: {csv_output}")
print(f" Cleaned JSON saved to: {json_output}")

# 7. Final price statistics summary

print("\n Final Average Prices Summary:")
summary = pd.DataFrame({
    'Dataset': ['CSV (Original)', 'JSON (Raw)', 'JSON (Adjusted)'],
    'Average Price': [
        round(csv_mean, 2),
        round(json_mean, 2),
        round(products_json['adjusted_price'].mean(), 2)
    ]
})
print(summary.to_string(index=False))
```

Post Cleaning Process:

1. Handling Missing Prices: The missing prices in the JSON datasets are filled using the median price from the available data.

This ensures that realistic and consistent values without distorting the data are achieved.

2. Validation Step- After imputation, the script checks that there are no remaining missing values, confirming that it now has a clean dataset ready for normalization.

3. Price Normalization- Because the CSV and JSON datasets may be from different sources, their price scales may differ. In this code, JSON prices are scaled according to a scaling factor calculated by the ratio of average prices in both datasets; this puts them on the same numerical scale. 4. Data Preservation: The two cleaned datasets are then saved into new files. This way, your raw data is left intact but you have ready-for-use standardized files for further analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Load cleaned datasets

csv = pd.read_csv("/products.csv")
# Load the previously cleaned JSON data which was saved as a CSV
json = pd.read_csv("/content/final_products_json.csv")

print(" Datasets loaded successfully!")

# Merge datasets on unique product IDs

if "Uniq_id" not in csv.columns or "uniq_id" not in json.columns:
    raise ValueError(" Missing 'Uniq_id' column in CSV or 'uniq_id' column in JSON file.")

# Ensure we use the adjusted price from the cleaned JSON data
if "Price" not in csv.columns or "adjusted_price" not in json.columns:
    raise ValueError(" Missing 'Price' in CSV or 'adjusted_price' in JSON.")

merged = csv.merge(
    json[["uniq_id", "adjusted_price"]],
    left_on="Uniq_id",
    right_on="uniq_id",
    how="inner"
)
print(f"Successfully merged: {len(merged)} records matched.")

# Calculate price differences
merged["price_diff"] = merged["Price"] - merged["adjusted_price"]

# Statistical summary

csv_mean = merged["Price"].mean()
json_mean = merged["adjusted_price"].mean()
avg_diff = merged["price_diff"].mean()

print("\n Price Comparison Summary:")
print(f"- CSV Average Price:   ${csv_mean:.2f}")
print(f"- JSON Average Price:  ${json_mean:.2f}")
print(f"- Average Difference:  ${avg_diff:.2f}")
print(f"- Median Difference:   ${merged['price_diff'].median():.2f}")
print(f"- Max Difference:      ${merged['price_diff'].abs().max():.2f}")

# Visualization: Scatter Plot

plt.figure(figsize=(8, 6))
plt.style.use("seaborn-v0_8-whitegrid")

# Scatter points
plt.scatter(merged["adjusted_price"], merged["Price"], alpha=0.6, color="steelblue", edgecolor="k", label="Product Prices")

# Perfect alignment line (y=x)
max_price = max(merged["Price"].max(), merged["adjusted_price"].max())
plt.plot([0, max_price], [0, max_price], 'r--', label="Perfect Alignment (y=x)")

# Add a regression line to show trend
m, b = np.polyfit(merged["adjusted_price"], merged["Price"], 1)
plt.plot(merged["adjusted_price"], m * merged["adjusted_price"] + b, color="green", linestyle="-", linewidth=1.5, label="Trend Line")

# Labels and title
plt.title("Price Comparison: CSV vs JSON", fontsize=14, pad=12)
plt.xlabel("JSON Adjusted Prices ($)", fontsize=12)
plt.ylabel("CSV Prices ($)", fontsize=12)
plt.legend(fontsize=10)
plt.tight_layout()

# Show plot
plt.show()

#Optional: Save merged comparison results
merged.to_csv("price_comparison_results.csv", index=False)
print("\n Saved comparison results to: price_comparison_results.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create price tier categories based on CSV prices

price_bins = [0, 50, 100, 200, 500, float('inf')]  # Tier boundaries
price_labels = ['<$50', '$50–100', '$100–200', '$200–500', '$500+']

merged['price_tier'] = pd.cut(
    merged['Price'],
    bins=price_bins,
    labels=price_labels,
    right=False  # Left-inclusive
)

# Confirm tier distribution
tier_counts = merged['price_tier'].value_counts().sort_index()
print("\n📊Price Tier Distribution:")
print(tier_counts)

#Set up visualization styling

plt.figure(figsize=(10, 6), dpi=120)
sns.set_theme(style="whitegrid", palette="Blues")


# Create boxplot (price differences by tier)


ax = sns.boxplot(
    data=merged,
    x='price_tier',
    y='price_diff',
    palette="Blues_r",     # Reverse palette for better tier contrast
    showfliers=False,      # Hide outliers for cleaner visualization
    width=0.65
)

# Add horizontal reference line at 0 (perfect parity)
plt.axhline(0, color='red', linestyle='--', linewidth=1, alpha=0.7, label="No Difference")

# Title, labels, and layout

plt.title("Price Differences by Product Price Tier", fontsize=15, pad=15)
plt.xlabel("Price Tier (based on CSV Prices)", fontsize=12, labelpad=8)
plt.ylabel("CSV Price − JSON Price ($)", fontsize=12, labelpad=8)
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right', frameon=True)

#Add median annotations to each box

for i, tier in enumerate(price_labels):
    tier_data = merged[merged['price_tier'] == tier]['price_diff'].dropna()
    if not tier_data.empty:
        median_val = tier_data.median()
        ax.text(
            i, median_val, f"{median_val:.2f}",
            ha='center', va='center',
            color='white', fontweight='bold',
            fontsize=9,
            bbox=dict(facecolor='black', alpha=0.5, boxstyle="round,pad=0.3")
        )
#Display the plot

plt.tight_layout()
plt.show()
```

```python
# Product Price Statistical Summary & Insights
import pandas as pd

# Summary Statistics
price_summary = products['Price'].describe()
mean, median = price_summary['mean'], price_summary['50%']

print("=== PRICE SUMMARY ===")
print(f"Mean:               ${mean:.2f}")
print(f"Median:             ${median:.2f}")
print(f"Min–Max Range:      ${price_summary['min']:.2f} → ${price_summary['max']:.2f}")
print(f"Standard Deviation: ${price_summary['std']:.2f}")
print(f"25th Percentile:    ${price_summary['25%']:.2f}")
print(f"75th Percentile:    ${price_summary['75%']:.2f}")

#
iqr = price_summary['75%'] - price_summary['25%']
price_range = price_summary['max'] - price_summary['min']
is_right_skewed = mean > median

print("\n===  DISTRIBUTION INSIGHTS ===")
print(f"Right-skewed:        {is_right_skewed} (Mean > Median)")
print(f"Total Price Range:   ${price_range:.2f}")
print(f"IQR (Middle Spread): ${iqr:.2f}")

# Outlier Detection
outlier_cutoff = price_summary['75%'] + 1.5 * iqr
outliers = (products['Price'] > outlier_cutoff).sum()

print("\n===  OUTLIER ANALYSIS ===")
print(f"Outlier Threshold:   > ${outlier_cutoff:.2f}")
print(f"Potential Outliers:  {outliers} products ({(outliers / len(products) * 100):.2f}%)")

#  Summary Note
print("\n Summary complete — use boxplot or histogram to visually confirm distribution shape.")
```

**LETS SEE HOW TO USE THE PRODUCTS URL'S**

```python
import pandas as pd
from IPython.display import Image, display
import requests

# Example DataFrame initialization for demonstration
data = {
    'product_name': ['Product 1'],
    'product_image_urls': [
        'https://share.google/lBhYtwgbUDPAfsM9d',  #URLs
    ]
}
products_df = pd.DataFrame(data)

def display_product_image(url):
    """Displays a product image from a URL."""
    try:
        # Ensure the URL is valid
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})

        # Check if the request was successful
        if response.status_code == 200:
            display(Image(url=url))  # Use the URL directly for display
        else:
            print(f"Failed to retrieve image; Status code: {response.status_code}")
    except Exception as e:
        print(f"Error displaying {url}: {e}")

# Loop through the DataFrame and display images
for index, row in products_df.iterrows():
    print(f"Displaying {row['product_name']}:")
    display_product_image(row['product_image_urls'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame initialization – replace with your actual data
data = {
    'product_name': ['Product A', 'Product A', 'Product B', 'Product B', 'Product C'],
    'review_score': [5, 4, 5, 3, 2],
    'review_text': [
        'Excellent product! Highly recommend.',
        'Very good, but could be improved.',
        'Absolutely love this! Will buy again.',
        'It’s okay, but not what I expected.',
        'Did not meet my expectations.'
    ]
}
reviews_df = pd.DataFrame(data)

# Calculate top reviews and best score
best_reviews = reviews_df[reviews_df['review_score'] == 5]

# Display best reviews
print("Best Reviews:")
print(best_reviews[['product_name', 'review_text']])

# Data Visualization
plt.figure(figsize=(10, 6))
sns.countplot(data=reviews_df, x='review_score', palette='viridis')
plt.title('Distribution of Review Scores')
plt.xlabel('Review Score')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.6)
plt.show()

# Additional Insights: Average score per product
average_scores = reviews_df.groupby('product_name')['review_score'].mean().reset_index()

# Plotting average scores per product
plt.figure(figsize=(10, 6))
sns.barplot(data=average_scores, x='product_name', y='review_score', palette='mako')
plt.title('Average Review Score per Product')
plt.xlabel('Product Name')
plt.ylabel('Average Review Score')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.6)
plt.show()
```

The Python code provided aims at comprehensively analyzing product reviews, especially in identifying the best reviews and visualization of data using plots. To begin with, it initializes a sample DataFrame, `reviews_df`, containing product names, review scores, and the corresponding review text. This is encouraged to be replaced by the user with their review data. Then, it filters for the best reviews, precisely those scoring a perfect 5, and prints them for quick reference. To put a deeper understanding into the dynamics of reviews, the code plots a visual using Matplotlib and Seaborn.

The first plot is the distribution of review scores, which enables the user to quickly make out a pattern in the customer feedback. Further, it plots a bar chart for the average review score each product gets to let one know which products are doing well on average.
